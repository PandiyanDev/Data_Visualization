### Scatterplot, coloring by labels

```python
fig, ax = plt.subplots()
# First, plot the data where the individual does not have heart disease
df_subset = df[df['target'] == 0]
ax.scatter(df_subset['trestbps'].values, df_subset['chol'].values, label = "healthy")

# second, plot the data where the individual does have heart disease
df_subset = df[df['target'] == 1]
ax.scatter(df_subset['trestbps'].values, df_subset['chol'].values, label = "not healthy")
ax.legend()
plt.show()
```

```python
label_mapping = {0:'healthy', 1: 'heart disease'}
```

```python
fig, ax = plt.subplots()
for target_value in df['target'].unique():
    df_subset = df[df['target'] == target_value]
    ax.scatter(df_subset['trestbps'].values, df_subset['chol'].values, label=label_mapping[target_value])

ax.legend()
plt.show()
```

### Scatterplot, calling out a single point

```python
### Bar chart with a line graph and a horizontal line

fig, ax = plt.subplots()
ax.plot(df_groupby_age.index, df_groupby_age['chol'].values, c='orange', label='chol')
ax.bar(df_groupby_age.index, df_groupby_age['trestbps'].values,label="trestbps")
ax.legend()
plt.show()
```

### Section #4 - Adding and altering text

##### Axis labels

```python
# bar chart
fig, ax = plt.subplots()
ax.bar(df_groupby_age.index, df_groupby_age['trestbps'].values,label="trestbps")
ax.set_xlabel("Age")
ax.set_ylabel("Average Cholestarol")
ax.set_title("Average Cholestarol by Age")
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12,10))
# First graph
axs[0, 0].bar(df_groupby_age.index, df_groupby_age['trestbps'].values)
axs[0, 0].set_xlabel("Age")
axs[0, 0].set_ylabel("Average Resting Blood Pressure")
axs[0, 0].set_title("Average Resting Blood Pressure by Age")

# Second graph
axs[0, 1].plot(df_groupby_age.index, df_groupby_age['chol'].values)
axs[0, 1].set_xlabel("Age")
axs[0, 1].set_ylabel("Average Cholesterol")
axs[0, 1].set_title("Average Cholesterol by Age")

# Third graph
axs[1, 0].scatter(df_groupby_age.index, df_groupby_age['thalach'].values)
axs[1, 0].set_xlabel("Age")
axs[1, 0].set_ylabel("Average Maximum Heart Rate Achieved")
axs[1, 0].set_title("Average Maximum Heart Rate Achieved by Age")

# Fourth graph
axs[1, 1].bar(df_groupby_age.index, df_groupby_age['oldpeak'].values)
axs[1, 1].set_xlabel("Age")
axs[1, 1].set_ylabel("Average Oldpeak")
axs[1, 1].set_title("Average Oldpeak by Age")

plt.show()
```

### Section #5 - common questions

#### Q. How do we change the figure size?

Here, we use figsize attribute to pass subplots 

```python
# smooth line oscillation
y_damped_oscillation = np.exp(-x) * np.cos(2 * np.pi * x)  
```

```python
fig, ax = plt.subplots()
ax.plot(x, y_damped_oscillation)
plt.show()
```

#### How do we save our image to a file?

- Here, we save image file using "savefig" syntax

```python
fig, ax = plt.subplots()
ax.plot(x, y_damped_oscillation)
plt.savefig("my-plot.png")
```

#### How do we change font size (title, axis labels and more..)?

```python
fig, ax = plt.subplots()

ax.plot(x, y_damped_oscillation)
ax.set_title("Damped Oscillation", fontsize= 15)
ax.set_xlabel("X label", fontsize = 12)
ax.set_ylabel("Y label", fontsize = 12)
ax.tick_params(axis="both", labelsize=10)

plt.show()
```

#### What the heck does this mean? `fig.add_subplot(111)`

1. First, fig.add_subplot() adds an axes to an existing figure.

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y_damped_oscillation)

ax2 = fig.add_subplot(2, 3, 3)
ax2.plot(x, y)
plt.show()
```

#### How do I change the tick frequency on the x or y axis?

```python
y = y_damped_oscillation 
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xticks([-10, -5, -1, 0 ,1, 5])
plt.show()
```

#### How do you set y-axis limit?

#### How do you plot two histograms on the same chart?

```python
# data points
dist1 = np.random.normal(loc=0, scale=1, size=1000)
dist2 = np.random.normal(loc=0.9, scale=0.5, size=1000)
```

```python
fig, ax = plt.subplots()
ax.hist(dist1, alpha =0.7, label = 'dist1')
ax.hist(dist2, alpha =0.7, label = 'dist2')
ax.legend()
plt.show()
```

#### How do you draw horizontal and vertical line?

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.hlines([-50, 50], min(x), max(x), colors = ['r', 'b'])
ax.vlines([-4], min(y), max(y), color='black')
plt.show()
```

```python
# let' s make data points
x_list = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y_list = [x*2 for x in x_list]
x_sorted =sorted(x_list, reverse=True)
colors = ['#EE6666', 'blue'] + ['green'] *(len(x) -2)
fig, ax = plt.subplots()
bars = ax.barh(x_sorted, y_list, color=colors)
# bar_clr = bars[0].get_facecolor()

# remove or alive line spine(box line)
ax.spines['right'].set_visible(True)
ax.spines['top'].set_visible(False)

# adding data points on each bar
for bar in bars:
    width = bar.get_width()
    ax.text(
        width+0.2,
        (bar.get_y() + bar.get_height() / 2), 
        str(width), 
        ha='left', 
        va='center',
        color = bar.get_facecolor()
    )
plt.show()
```

```python
# let' s make data points
def bar_and_lines():
    x_list = [1, 2, 3, 4, 5, 6, 7, 8, 9]
    y_list = [x*2 for x in x_list]
    x_sorted =sorted(x_list, reverse=True)
    colors = ['#EE6666', 'yellow'] + ['green'] *(len(x) -2)
    fig, ax = plt.subplots(1, 2, figsize=(11, 4))
    ax[0].barh(x_sorted, y_list, color=colors)
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(['a','b', 'c'])
    plt.show()
```

```python
bar_and_lines()
```

```python
with plt.style.context('ggplot'):
    bar_and_lines()
```

### list of plot style and corresponding plots

```python
for i in plt.style.available:
    print(i)
    with plt.style.context(i):
        bar_and_lines()
```
